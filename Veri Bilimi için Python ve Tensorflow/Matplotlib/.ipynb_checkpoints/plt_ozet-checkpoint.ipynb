```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [10,20,30,30,30,40,50,60,70,75]
```

```python
weightList = [20,60,70,80,86,87,70,90,95,90]
```

```python
# We can do numerical operations on lists
ageArray = np.array(ageList)
weightArray = np.array(weightList)
```

```python
# 'r' means red line, xlabel sets name of x axis, ylabel sets name of y axis
plt.plot(ageArray, weightArray, 'r x')
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age/Weight Correlation Table")
plt.show()
```

```python
arr1 = np.linspace(0,10,20)
arr2 = arr1 ** 3
```

```python
plt.plot(arr1, arr2, 'g +')
```

```python
plt.plot(arr1, arr2, 'b--')
```

```python
# plt.subplot(m, n) : m = how many rows, n: how many columns
plt.subplot(1,2,1)
plt.plot(arr1, arr2, "r*-")
plt.subplot(1,2,2) # 1.sıra, 2.kolon. 2.grafik
plt.plot(arr2,arr1, "g--")
```

```python
# add_axes tuple (left, bottom, width, height)
myFig = plt.figure()
figureAxes = myFig.add_axes([0.8, 0.8, 0.6, 0.6])
figureAxes.plot(arr1, arr2, 'g--')
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
## İç içe grafikler ( Nested plots )
fig2 = plt.figure()
axis1 = fig2.add_axes([0.1, 0.1, 0.9, 0.9])
axis2 = fig2.add_axes([0.2, 0.6, 0.3, 0.3])

axis1.plot(arr1, arr2, "g")
axis1.set_xlabel("X Ekseni")
axis1.set_ylabel("Y Ekseni")
axis1.set_title("Ana Grafik Başlık")

axis2.plot(arr2, arr1, "b--")
axis2.set_xlabel("X Ekseni Küçük")
axis2.set_ylabel("Y Ekseni Küçük")
axis2.set_title("Küçük Grafik başlık")
```

```python
(myFig, myAxis) = plt.subplots(nrows=1, ncols=2)
# type(myAxis) numpy.ndarray
for axe in myAxis:
    axe.plot(arr1, arr2, "g")
    axe.set_xlabel("X Ekseni")

# Grafikler arasındaki sıkışıklığı önlüyor
plt.tight_layout()
```

```python
# dpi bulanıklık, kaliteyi
newFig = plt.figure(dpi = 100)

newAxis = newFig.add_axes([0.1, 0.1, 0.9, 0.9])
newAxis.plot(arr1, arr1 ** 2, label = "arr1 ** 2")
newAxis.plot(arr1, arr1 ** 3, label = "arr1 ** 3")
# Labeling arr ** 2 and arr1 ** 3
newAxis.legend()
```

```python
newFig.savefig("myfigure.png", dpi=100)
```
