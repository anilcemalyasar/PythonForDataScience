```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [10,20,30,30,30,40,50,60,70,75]
```

```python
weightList = [20,60,70,80,86,87,70,90,95,90]
```

```python
# We can do numerical operations on lists
ageArray = np.array(ageList)
weightArray = np.array(weightList)
```

```python
# 'r' means red line, xlabel sets name of x axis, ylabel sets name of y axis
plt.plot(ageArray, weightArray, 'r x')
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age/Weight Correlation Table")
plt.show()
```
